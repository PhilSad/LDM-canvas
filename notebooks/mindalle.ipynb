```python
from min_dalle import MinDalle
import torch 
```

```python
model = MinDalle(
    models_root='./pretrained',
    dtype=torch.float16,
    device='cuda',
    is_mega=True, 
    is_reusable=True
)
```

```python
image = model.generate_image(
    text='Nuclear explosion broccoli',
    seed=-1,
    grid_size=1,
    is_seamless=False,
    temperature=1,
    top_k=256,
    supercondition_factor=32,
    is_verbose=False
)
```
